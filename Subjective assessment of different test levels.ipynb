```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data from JSON files into pandas DataFrames
data_default = pd.read_json(r'C:\Users\Addy\OneDrive\Dokumente\Subjective_Results\Rattings\Rattings_default.json')
data_low = pd.read_json(r'C:\Users\Addy\OneDrive\Dokumente\Subjective_Results\Rattings\Rattings_low.json')
data_mid = pd.read_json(r'C:\Users\Addy\OneDrive\Dokumente\Subjective_Results\Rattings\Rattings_mid.json')
data_high = pd.read_json(r'C:\Users\Addy\OneDrive\Dokumente\Subjective_Results\Rattings\Rattings_high.json')
```

```python
#Drop the unneccassary columns
data_default = data_default.drop(['_id', '__v' , 'Q5_text' , 'Q6_text'], axis=1)
data_low = data_low.drop(['_id', '__v' , 'Q5_text' , 'Q6_text'], axis=1)
data_mid = data_mid.drop(['_id', '__v' , 'Q5_text' , 'Q6_text'], axis=1)
data_high = data_high.drop(['_id', '__v' , 'Q5_text' , 'Q6_text'], axis=1)


```

```python

# Add the title "Participants" to the index column
data_default = data_default.rename_axis("Partcipants")
data_low = data_low.rename_axis("Partcipants")
data_mid = data_mid.rename_axis("Partcipants")
data_high = data_high.rename_axis("Partcipants")
```

```python

# Add the title "Participants" to the index column
data_default.index = data_default.index + 1
data_low.index = data_low.index + 1
data_mid.index = data_mid.index + 1
data_high.index = data_high.index + 1

data_high

```

```python
# Define the mapping of categorical labels to numerical values
rating_mapping = {'Bad': 1, 'Poor': 2, 'Fair': 3, 'Good': 4, 'Excellent': 5,"Extreme effort" :1, "Considerable effort" :2,
                  "Moderate effort" :3,"Minor effort" :4 , "No effort": 5,
                  'Yes': 1, 'No': 0}

# Convert the categorical labels to numerical values
data_default = data_default.replace(rating_mapping)
data_low = data_low.replace(rating_mapping)
data_mid = data_mid.replace(rating_mapping)
data_high = data_high.replace(rating_mapping)

```

```python
# Caluclating MOS or mean square

mos_overall_default = data_default.mean()
mos_overall_low = data_low.mean()
mos_overall_mid = data_mid.mean()
mos_overall_high = data_high.mean()
```

```python

# Visualisation Overall MOS

mos_scores = [mos_overall_default.values[0], mos_overall_low.values[0], mos_overall_mid.values[0], mos_overall_high.values[0]]


conditions = ['Default', 'Low', 'Mid', 'High']


plt.bar(conditions, mos_scores)
plt.xlabel('Test Condition')
plt.ylabel('Overall MOS')
plt.title('Overall MOS Comparison')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Retrieve all values of Q5 from the dataset first for each test level
q5_values_default = data_default['Q5']
q5_values_low = data_low['Q5']
q5_values_mid = data_mid['Q5']
q5_values_high = data_high['Q5']

# Replace 1 and 0 with "Yes" and "No" in the Q5 values
q5_values_default_replaced = ["Yes" if val == 1 else "No" for val in q5_values_default]
q5_values_low_replaced = ["Yes" if val == 1 else "No" for val in q5_values_low]
q5_values_mid_replaced = ["Yes" if val == 1 else "No" for val in q5_values_mid]
q5_values_high_replaced = ["Yes" if val == 1 else "No" for val in q5_values_high]

# Combine the replaced Q5 values into dictionaries
data_default_q5 = {
    'Q5': q5_values_default_replaced
}

data_low_q5 = {
    'Q5': q5_values_low_replaced
}

data_mid_q5 = {
    'Q5': q5_values_mid_replaced
}

data_high_q5 = {
    'Q5': q5_values_high_replaced
}

# Combine the datasets into dictionaries
datasets_q5 = {
    'Default': data_default_q5,
    'Low': data_low_q5,
    'Mid': data_mid_q5,
    'High': data_high_q5
}

# Initialize a list to store the pie chart labels and counts
labels_q5 = []
counts_q5 = []

# Loop through each category (Default, Low, Mid, High) for Q5
for category, dataset in datasets_q5.items():
    # Create a DataFrame for the current category's dataset for Q5
    df = pd.DataFrame({'Q5': dataset['Q5']})
    
    # Count the number of "Yes" and "No" in the 'Q5' column
    value_counts = df['Q5'].value_counts()
    
    # Append the category name to labels for Q5
    labels_q5.append(f'{category} (Q5 - {len(df)})')
    
    # Append the value counts to counts for Q5
    counts_q5.append(value_counts)

# Create subplots to display multiple pie charts for Q5
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Did you have any difficulty in communicating? ')

# Plot pie charts for Q5 for each category
for i, ax in enumerate(axes.flat):
    ax.pie(counts_q5[i], labels=counts_q5[i].index,
           autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightcoral'])
    ax.set_title(labels_q5[i])
    ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Adjust the layout and display the pie charts
plt.tight_layout()
plt.subplots_adjust(top=0.9)  # Adjust the suptitle position
plt.show()
```
